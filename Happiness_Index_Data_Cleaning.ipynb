***As We Can See That Region Column Is Missing In 2017 Dataset, We Can Insert A Region Column Using The 2015 or 2016 Dataset As There Are Almost Every Country Of 2017 Is Present In The 2015 or 2016 Dataset. So to do this :-***

```python
df_2017.head()
```
